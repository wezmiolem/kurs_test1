```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data

mean_vals = np.mean(X, axis=0)
std_vals = np.std(X, axis=0)

X_standardized = (X - mean_vals) / std_vals

scaler = StandardScaler()
X_sklearn = scaler.fit_transform(X)

mean_after = np.mean(X_standardized, axis=0)
std_after = np.std(X_standardized, axis=0)
max_diff = np.max(np.abs(X_standardized - X_sklearn))

print(f"Matrix shape: {X_standardized.shape}")

print(f"Mean: {mean_after}")
print(f"Std: {std_after}")

print(f"\nComparison with Sklearn:")
print(f"Max difference: {max_diff}")

if np.allclose(X_standardized, X_sklearn):
    print("\noutcome is the same")

```
