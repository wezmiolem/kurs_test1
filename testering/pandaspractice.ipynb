Wczytaj dataset Titanic i oblicz średni wiek pasażerów według klasy ( Pclass ) i płci ( Sex ).
Dataset: Titanic
(https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv)
Wymagania:
Oczekiwany rezultat:
(proste)
Użyj .groupby() z dwoma kolumnami
Oblicz średnią wieku ( .mean() )
Wyświetl wyniki w formie czytelnej tabeli
Tabela ze średnim wiekiem dla 6 kombinacji (3 klasy × 2 płcie)

```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

avg_age = titanic.groupby(['Pclass', 'Sex']).agg({'Age': ['mean']})

print(avg_age)
```

Dla datasetu Titanic policz, ile pasażerów było w każdej klasie i czy przeżyli ( Survived ).
Dataset: Titanic
Wymagania:
Oczekiwany rezultat:
(proste)
Grupuj według Pclass i Survived
Użyj .size() lub .count()
Przedstaw wyniki jako tabelę pivot
Tabela pokazująca liczbę pasażerów dla każdej kombinacji klasy i statusu przeżycia

```python
titanic2 = titanic.copy()
survivors = titanic2.groupby(['Pclass', 'Survived'])['Survived'].count().unstack()
survivors2 = titanic2.pivot_table(
    index='Pclass', 
    columns='Survived',  
    aggfunc='size'
)
print(survivors)
print('')
print(survivors2)
```

```python
np.random.seed(42)
sales = pd.DataFrame({
'Region': np.random.choice(['A', 'B', 'C'], 100),
'Sprzedaż': np.random.randint(50, 500, 100)
})

stats_by_region = sales.groupby(['Region'])['Sprzedaż'].agg(['min', 'max', 'mean', 'std']).round(2)
print(stats_by_region)
```

Napisz własną funkcję, która oblicza zakres międzykwartylowy (IQR = Q3 - Q1) i zastosuj ją
do grupy.
Dataset: California Housing
Wymagania:
Oczekiwany rezultat:
(średnie)
Napisz funkcję iqr_func(x) obliczającą IQR
Zastosuj ją do MedHouseVal według regionów
Porównaj IQR między regionami
Tabela z IQR dla każdego regionu

```python
import numpy as np
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing(as_frame=True)
df_housing = housing.frame

def assign_region(row):
    lat, lon = row['Latitude'], row['Longitude']
    if lat > 37.5 and lon < -121:
        return 'Bay_Area'
    elif lat < 35 and lon > -119:
        return 'SoCal'
    elif -121 <= lon <= -118:
        return 'Central_Valley'
    else:
        return 'North_Inland'
    

df_housing['region_name'] = df_housing.apply(assign_region, axis=1)


def calc_iqr(dane):
    q1 = np.percentile(dane, 25)
    q3 = np.percentile(dane, 75)
    return q3 - q1

iqr_results = df_housing.groupby('region_name')['MedHouseVal'].agg(calc_iqr).reset_index()
iqr_results.columns = ['Region', 'IQR_MedHouseVal']
print(iqr_results)
```

Dla datasetu California Housing znormalizuj MedInc (dochód) w każdym regionie do
zakresu [0, 1].
Dataset: California Housing
Wymagania:
Oczekiwany rezultat:
(średnie)
Stwórz regiony (np. na podstawie Latitude > median)
Napisz funkcję normalizacji Min-Max
Użyj .groupby().transform() lub .apply()
Zweryfikuj, że min=0 i max=1 w każdym regionie
DataFrame z kolumną znormalizowaną grupowo

```python
def min_max_normalize(series):
    return (series - series.min()) / (series.max() - series.min())

df_housing['MedInc_Normalized'] = df_housing.groupby('region_name')['MedInc'].transform(min_max_normalize)
print(df_housing)
```

```python
region_minmax = df_housing.groupby(['region_name'])['MedInc_Normalized'].agg(['min', 'max']).reset_index()
print(region_minmax)
```
